### Q1. One‑Sample t‑Test (Mean)
A company claims that the average working hours of employees is **8 hours/day**. A sample of **25 employees** shows a mean of **7.6 hours** with a standard deviation of **0.8 hours**. Test the claim at **5% significance level**.

**Hint:** Population standard deviation is unknown.

```python
import numpy as np 
import scipy.stats as stats 

population_mean = 8 
n = 25 
sample_mean = 7.6 
stdv = 0.8 
alpha = 0.05 
df = n - 1 
t_stats = (sample_mean- population_mean)/ stdv/np.sqrt(n)

p_value = 2 * (1 - stats.t.cdf (abs(t_stats),df))

if p_value<alpha:
    print("null hypothesis, company claim is true")
else:
    print("alternate hypothesis, company claim is false")

```

### Q2. One‑Sample Z‑Test (Proportion)
A website claims that **70% users** like its new UI. In a survey of **200 users**, **130 liked it**. Test the claim at **5% significance level**.

**Hint:** Convert to sample proportion.

```python
n = 200 
Sample_proportion = 0.65
Claimed_proportion = 0.70 
alpha = 0.05

z_stat = (Sample_proportion - Claimed_proportion)/ np.sqrt(Claimed_proportion * (1-Claimed_proportion)/n)

p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

if p_value<alpha:
    print("Alternate hypthesis, claimed is false")
else:
    print("Null hypothesis, claimed is correct")
```

### Q3. Two‑Sample t‑Test (Independent)
Two teaching methods are compared.
- Method A: mean score = **75**, SD = **6**, n = **30**
- Method B: mean score = **72**, SD = **5**, n = **28**

Is there a significant difference in mean scores?

**Hint:** Assume equal variances.

```python
mean1 = 75 
mean2 = 72
sd1 = 6
sd2 = 5 
n1 = 30
n2 = 28
df = (n1+n2-2)

t_stats = (mean1 - mean2) / np.sqrt(sd1/n1+sd2/n2)
p_value = 2 * (1 - stats.t.cdf(abs(t_stats),df))

alpha = 0.05 

if p_value<alpha:
    print("reject null Hypothesis, there is a difference")
else:
    print("accept null Hypothesis, There is no difference")
```

### Q4. Paired t‑Test
Weights (kg) of 8 people were recorded **before and after** a diet program.
Before: [78, 82, 85, 80, 76, 90, 88, 84]
After:  [75, 80, 83, 78, 74, 87, 85, 82]

Test whether the program is effective.

**Hint:** Work on differences.

```python
Before = np.array([78, 82, 85, 80, 76, 90, 88, 84])
After = np.array([75, 80, 83, 78, 74, 87, 85, 82])

alpha = 0.05 

t_stats, p_value = stats.ttest_rel(Before,After)

print("P-Value:", p_value)
print("T-Stats:", t_stats)


if p_value<alpha:
    print("rejecting Null hypothesis, Program is effective")
else:
    print("not rejecting Null hypothesis, Program is not effective")
```

### Q5. Chi‑Square Test (Goodness of Fit)
A dice is rolled **120 times**. The observed frequencies are:
[15, 18, 22, 20, 25, 20]

Test whether the dice is fair.

**Hint:** Expected frequency = total / 6.

```python
observed = np.array([15, 18, 22, 20, 25, 20])
expected = np.array([20,20,20,20,20,20])

alpha = 0.05
import numpy as np 
from scipy import stats

chi_stat,p_value = stats.chisquare(observed,expected)

print("Chi", chi_stat)
print("P-value", p_value)

if p_value<alpha:
    print("rejecting Null hypothesis, dice is unfair")
else:
    print("not rejecting Null hypothesis, fair dice")
```
